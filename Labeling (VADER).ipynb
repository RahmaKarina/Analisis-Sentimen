```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
```

### MEMBERIKAN LABEL

```python
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["teks"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["teks"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["teks"]]
df['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in df["teks"]]
df.head()
```

```python
score = df["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
df["Sentiment"] = sentiment
df.head()
```
