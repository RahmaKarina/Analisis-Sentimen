# CLEANING

## Menghapus retwett, username, tag

```python
text_clean = []

for idx, text in enumerate(df_new['teks']):
    clean_text = str(text).replace(str(df_new['username'][idx]), '')
    clean_text = re.sub('@[\w]+','',clean_text)
    clean_text = re.sub("b'@",'',clean_text)
    clean_text = re.sub("b'RT :",'',clean_text)
    clean_text = re.sub("b'",'',clean_text)
    clean_text = re.sub('b"','',clean_text)
    text_clean.append(clean_text)
    
df_new['Text_Clean'] = text_clean
```

```python
df_new['Text_Clean'] = df_new['Text_Clean'].drop_duplicates()
df_new = df_new.dropna()
df_new = df_new.reset_index(drop=True)
df_new
```

# CASE FOLDING

### Menghapus spasi, https, angka, tanda baca, link

```python
def Case_Folding(text):
    #menghapus link
    text = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", text)
    # Menghapus Tanda Baca
    text = re.sub(r'[^\w]|_',' ', text)
    
    # Menghapus Angka
    text = re.sub("\S*\d\S*", "", text).strip()
    text = re.sub(r"\b\d+\b", " ", text)
    
    # Mengubah text menjadi lowercase
    text = text.lower()
    
    # Menghapus white space
    text = re.sub('[\s]+', ' ', text)
    text = text.replace('\n',' ')
    
    return text
```

```python
df_new['Case_Folding'] = df_new['Text_Clean'].apply(Case_Folding)
df_new.head()
```

# TOKENIZING

### Mengubah menjadi token

```python
def split_word(teks):
    list_teks = []
    for txt in teks.split(" "):
        list_teks.append(txt)
    return list_teks
```

```python
df_new['Tokenizing'] = df_new['Case_Folding'].apply(split_word)
df_new.head()
```

# NORMALISASI

### Mengubah slang

```python
slang_dictionary = pd.read_csv('colloquial-indonesian-lexicon-1.csv')
slang_dict = pd.Series(slang_dictionary['formal'].values,index=slang_dictionary['slang']).to_dict()
```

```python
def normalisasi(text):
    for word in text:
        if word in slang_dict.keys():
            text = [s.replace(word, slang_dict[word]) for s in text]
    return text
```

```python
df_new['Normalisasi'] = df_new['Tokenizing'].apply(normalisasi)
df_new.head(50)
```

# STOPWORD REMOVAL

### Mengubah kata yang tidak perlu dan menghapus kata bahasa inggris

```python
def remove_stopwords(token):
    text = [word for word in token if word not in listStopword]
    return text
```

```python
def remove_stopwords2(teks):
    unwanted_words = ['https', 'google','kamu','halo','ga', 'gue', 'lo','xef', 'lu', 'dan','saya','dia','c','sebelumgojek','xef', 'xae', 'xbb', 'xae', 'xae', 'xaa', 'xae', 'xbb', 'xae','xbd', 'xef', 'ig', 'bts', 'nhttps','xed', 'xec', 'xeb', 'xab', 'xeb', 'btsv', 'nhttps','aklsdjlsakjduluakdj','aaa','anjing','instagram','youtube','tiktok','shopee','coba','pvc','xxl','wts','sebekamumgojek','wts','wtb','voc','deh','tidak','cek','kak','sih','nya','pc','photocard','renjun','ntags','gc','kya','nsh','mah','lol','heh','allah','hai','xcb','bf','xp','nwtb','yeay','nsiwon','nct','jaehyun','eajpark','zonauang','kabs','jaemin','nder','kfc','spf','npreloved','promo','voc','nfree','dieng','loh','haechan','tags','idc','photocards','hahaha','tfl','xba','xb','plsj','twt','mcd']
    text = [word for word in teks if not word in unwanted_words]
    return text
```

```python
def remove_english(teks):
  text = [i for i in teks if i not in words.words()]
  return text
```

```python
df_new['Stopword'] = df_new['Normalisasi'].apply(lambda x: remove_stopwords(x))
df_new['Stopword English'] = df_new['Stopword'].apply(lambda x: remove_english(x))
df_new['Stopword English'] = df_new['Stopword English'].apply(lambda x: remove_stopwords2(x))

df_new.head(20)
```

# STEMMING

### Menghilangkan kata imbuhan

```python
# Membuat stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def stem(teks):
  text = [stemmer.stem(word) for word in teks]
  return text
```

```python
df_new['Stemming'] = df_new['Stopword English'].apply(lambda x: stem(x))
df_new.head()
```
