```python
import pandas as pd 
import numpy as np 
```

```python
df_new = pd.read_csv("Hasil Label-Preproc - Versi uji 2.csv")
df_new.shape
```

```python
df_new
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Polarity == 0 positive
train_s1 = df_new[df_new["Sentiment"] == 'Positive']
all_text_s1 = ' '.join(word for word in train_s1["Stemming"])
wordcloud = WordCloud(width=1000, height=1000, colormap='Blues', background_color='black', mode='RGBA').generate(all_text_s1)
plt.figure( figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()

# Polarity == 1 negatif
train_s2 = df_new[df_new["Sentiment"] == 'Negative']
all_text_s2 = ' '.join(word for word in train_s2["Stemming"])
wordcloud = WordCloud(width=1000, height=1000, colormap='Reds', background_color='black', mode='RGBA').generate(all_text_s2)
plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
